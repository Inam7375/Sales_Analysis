```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

Yearly_Sales['City'] = Yearly_Sales['Purchase Address'].apply(lambda x: get_city(x)+' '+get_state(x))
Yearly_Sales.head()
```

```python
results = Yearly_Sales.groupby('City').sum()
```

```python
cities = [city for city, df in Yearly_Sales.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=9)
plt.ylabel('Sales in USD $')
plt.xlabel('Month Name')
plt.show()
```

## What time should we display advertisements to maximize the likelihood of customer's byuing product?

```python
Yearly_Sales['Order Date'] = pd.to_datetime(Yearly_Sales['Order Date'])
Yearly_Sales.head()
```

```python
Yearly_Sales['Hours'] = Yearly_Sales['Order Date'].dt.hour
Yearly_Sales['Minutes'] = Yearly_Sales['Order Date'].dt.minute
Yearly_Sales.head()
```

```python
hours = [hour for hour, df in Yearly_Sales.groupby('Hours')]

plt.plot(hours, Yearly_Sales.groupby(['Hours']).count())
plt.xticks(hours)
plt.grid()
plt.ylabel('Number of Products')
plt.xlabel('Hours')
plt.show()
```
